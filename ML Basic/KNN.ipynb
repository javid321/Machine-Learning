```python
feature_df = pd.DataFrame(features.tolist(), columns=data.feature_names.tolist())
feature_df.head()
```

```python
target_df = pd.DataFrame(data.target.tolist(), columns=['target'])
target_df.value_counts()
```

```python
feature_df.shape
```

```python
target_df.shape
```

```python
pd.set_option('display.max_columns', 31)
data_df = pd.concat([feature_df, target_df], axis=1)
data_df
```

```python
data_df.info()
```

```python
X_data = data_df.drop(['target'], axis=1)
```

```python
y_data = data_df['target']
```

```python
X_train, x_test, y_train, y_test = train_test_split(X_data, y_data, test_size=0.2)
```

```python
print(X_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python
model_KNN = KNeighborsClassifier(n_neighbors=3)
```

```python
model_KNN.fit(X_train, y_train)
```

```python
y_test
```

```python
x_test
```

```python
y_pred = model_KNN.predict(x_test)
y_pred
```

```python
print(model_KNN.score(x_test, y_test))
```
