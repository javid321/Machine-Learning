```python
import cv2
```

```python
cap = cv2.VideoCapture('DonaldBeanDeepfake.mp4')
fps = cap.get(cv2.CAP_PROP_FPS)
face_clf = cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')

while True:
    succes, frame = cap.read()
    if succes == False:
        break
    frame_gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_clf.detectMultiScale(frame_gray, scaleFactor=1.1, minNeighbors=5)
    for (x, y, w, h) in faces:
        print(x, y, w, h)
        cv2.rectangle(frame, (x,y), (x+w, y+h), (0, 255, 0), 5)
    cv2.imshow('frame', frame)
    if cv2.waitKey(5) == ord('m'):
        break

cap.release()
cv2.destroyAllWindows()
```
