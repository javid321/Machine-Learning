# Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataclean_df.drop(columns=['date', 'snow_depth'])
y = dataclean_df['snow_depth']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
model_dtree = DecisionTreeRegressor()
model_rforest = RandomForestRegressor()
```

```python
model_dtree.fit(X_train, y_train)
model_rforest.fit(X_train, y_train)
```

```python
model_dtree.score(X_train, y_train), model_dtree.score(X_test, y_test)
```

```python
model_rforest.score(X_train, y_train), model_rforest.score(X_test, y_test)
```

```python
from sklearn.pipeline import Pipeline
```

```python
model_pipeline=Pipeline([
    ('model_clf', RandomForestRegressor(n_jobs=-1))
])
```

# GidSearcV

```python
from sklearn.model_selection import GridSearchCV
```

```python
model_pipeline.get_params()
```

```python
parameter = {
    'model_clf__n_estimators': [100, 150],
    'model_clf__max_depth': [20, 50, 80],
    'model_clf__max_features' : [0.3, 0.6, 0.8],
    'model_clf__min_samples_leaf':[1, 5, 10]
}
```

```python
model_grid = GridSearchCV(model_pipeline, parameter, cv= 3, n_jobs=-1, verbose=1)
```

```python
model_grid.fit(X_train, y_train)
```

```python
pd.DataFrame(model_grid.cv_results_).sort_values('rank_test_score')
```
